```python
import tensorflow as tf
from pathlib import Path
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option("display.max_columns",100)
```

```python
df = pd.read_csv('training_set_features.csv',
                index_col="respondent_id")
df.head()
```

```python
df2 = pd.read_csv('training_set_labels.csv')
df2.head()
df2.info()
```

```python
dfs = []
dfs.append(df)
dfs.append(df2)
df3 = pd.concat(dfs, axis = 1)
df3 = df3.drop(['respondent_id'],axis=1)
```

```python
df3.head()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df3.corr(),cmap='Blues',annot=False) 
```

```python
sns.heatmap(df3.isnull(),yticklabels=False,cmap = 'viridis')
```

```python
#Quality correlation matrix
k = 12 #number of variables for heatmap
cols = df3.corr().nlargest(k, 'h1n1_vaccine')['h1n1_vaccine'].index
cm = df3[cols].corr()
plt.figure(figsize=(10,6))
sns.heatmap(cm, annot=True, cmap = 'viridis')
```
